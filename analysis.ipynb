```python
def get_cut_off(sorted_true, fake, percentile):
    cutoff = sorted_true[int(len(sorted_true) * percentile)]
    fake_remaining = len(fake[fake < cutoff]) / len(fake)
    true_remaining = len(sorted_true[sorted_true < cutoff]) / len(sorted_true)
    return cutoff, fake_remaining, true_remaining
    
```

```python
percent_remaining_by_region = [[] for i in range(25)]
for i in range(25):
    linear_true_remaining = linear_num_trues_by_region[i] / nocuts_num_trues_by_region[i]
    for p in np.linspace(linear_true_remaining - 0.05, 1.0, 500, endpoint=False):
        percent_remaining_by_region[i].append(get_cut_off(nocuts_rzChiSquared_true_by_region[i], nocuts_rzChiSquared_fake_by_region[i], p))
        
```

```python
import ipywidgets as widgets
from ipywidgets import interact, Layout
from IPython.display import display, HTML

# i = 14
# chosen = 450

def plot_eff_fake_curve(region, chosen):
    fake_remaining = [percent_remaining_by_region[region][j][1] * 100 for j in range(500)]
    true_remaining = [percent_remaining_by_region[region][j][2] * 100 for j in range(500)]
    plt.plot(fake_remaining, true_remaining, color="orange", label="possible helix cuts", zorder=1)
    linear_fake_remaining = linear_num_fakes_by_region[region] / nocuts_num_fakes_by_region[region] * 100
    linear_true_remaining = linear_num_trues_by_region[region] / nocuts_num_trues_by_region[region] * 100
    plt.scatter(linear_fake_remaining, linear_true_remaining, color="red", label="linear cuts", marker=".")
    plt.scatter(percent_remaining_by_region[region][chosen][1] * 100, percent_remaining_by_region[region][chosen][2] * 100, color="green", label="chosen helix cuts", marker=".")
    plt.xlabel("% fake remaining (100% = " + str(nocuts_num_fakes_by_region[region]) + ")")
    plt.ylabel("% true remaining (100% = " + str(nocuts_num_trues_by_region[region]) + ")")
    plt.title("Region " + str(region) + " (" + region_components[region] + ")")
    plt.legend()
    helix_cut = percent_remaining_by_region[region][chosen][0]
    print("Chosen cutoff: ", helix_cut)
    
    num_trues_remained_helix = len(nocuts_rzChiSquared_true_by_region[region][nocuts_rzChiSquared_true_by_region[region] < helix_cut])
    num_fakes_remained_helix = len(nocuts_rzChiSquared_fake_by_region[region][nocuts_rzChiSquared_fake_by_region[region] < helix_cut])
    num_partOfPT5_remained_helix = len(nocuts_rzChiSquared_partOfPT5_by_region[region][nocuts_rzChiSquared_partOfPT5_by_region[region] < helix_cut])
    
    cell_text = [["", "no cuts", "linear", "helix"],
                ["# trues", nocuts_num_trues_by_region[region], linear_num_trues_by_region[region], num_trues_remained_helix],
                ["# fakes", nocuts_num_fakes_by_region[region], linear_num_fakes_by_region[region], num_fakes_remained_helix],
                ["# part of pT5", nocuts_num_partOfPT5_by_region[region], linear_num_partOfPT5_by_region[region], num_partOfPT5_remained_helix]
                ]
    
    cell_colors = [["white", "white", "white", "white"],
                   ["white", "white", "white", "white"],
                   ["white", "white", "white", "white"],
                   ["white", "white", "white", "white"]
                   ]
    
    if num_trues_remained_helix < linear_num_trues_by_region[region]:
        cell_colors[1][3] = "tomato"
    else:
        cell_colors[1][3] = "lime"
        
    if num_fakes_remained_helix < linear_num_fakes_by_region[region]:
        cell_colors[2][3] = "lime"
    else:
        cell_colors[2][3] = "tomato"
        
    if num_partOfPT5_remained_helix < linear_num_partOfPT5_by_region[region]:
        cell_colors[3][3] = "tomato"
    else:
        cell_colors[3][3] = "lime"
        
    
    plt.table(cell_text, bbox=[0, -0.4, 1, 0.2], cellLoc="center", cellColours=cell_colors)
    
    # print("# part of pT5 total: ", nocuts_num_partOfPT5_by_region[region])
    
interact(
    plot_eff_fake_curve,
    region=widgets.IntSlider(min=0, max=24, step=1, value=0, Layout=Layout(width="200px")),
    chosen=widgets.IntSlider(min=0, max=499, step=1, value=400, Layout=Layout(width="200px"))
)
```
