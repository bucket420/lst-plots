```python
def get_cut_off(sorted_true, fake, percentile):
    cutoff = sorted_true[int(len(sorted_true) * percentile)]
    fake_remaining = len(fake[fake < cutoff]) / len(fake)
    true_remaining = len(sorted_true[sorted_true < cutoff]) / len(sorted_true)
    return cutoff, fake_remaining, true_remaining
    
```
